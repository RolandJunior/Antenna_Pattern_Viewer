```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import antenna pattern file
antenna = 'OA40-67-T8'
df = pd.read_csv(f'{antenna}.adf', sep=':', names=['Field','Data'])

# Subset vertical and horizontal patterns
vert = df['Field'].iloc[30:390]
horiz = df['Field'].iloc[394:754]

# Split data and adjust type
vert_angle = np.radians(list(vert.str.split(',', expand=True)[0].astype(int)))
vert_value = list(vert.str.split(',', expand=True)[1].astype(float))
horiz_angle = np.radians(list(horiz.str.split(',', expand=True)[0].astype(int)))
horiz_value = list(horiz.str.split(',', expand=True)[1].astype(float))

# Plot patterns
fig = plt.figure(figsize=(18,8))
fig.suptitle(antenna, fontsize=20)
ax1 = plt.subplot(121, projection='polar')  #121: 1 line, 2 columns, subplot 1
ax2 = plt.subplot(122, projection='polar')
ax1.set_title('Vertical', fontsize=18)
ax2.set_title('Horizontal', fontsize=18)
ax1.plot(vert_angle, vert_value, linewidth=2)
ax2.plot(horiz_angle, horiz_value, linewidth=2)
plt.show()
```
